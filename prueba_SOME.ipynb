```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Configure histogram
x_1 = range(0,13,1)
y_1 = list(np.random.randint(low=40, high=60, size=11))
y_1.append(50)
y_1.append(65)

x_2 = range(0,13,1)
y_2 = list(np.random.randint(low=5, high=15, size=12))
y_2.append(60)

x_3 = range(0,13,1)
y_3 = list(np.random.randint(low=0, high=30, size=11))
y_3.append(0)
y_3.append(94)

x_4 = range(0,13,1)
y_4 = list(np.random.randint(low=0, high=10, size=12))
y_4.append(25)


plt.figure(figsize=(16,8))

plt.plot(x_1, y_1, label='Madres', linewidth=1, color='red')
plt.plot(x_2, y_2, label='Hombres que regalan', linewidth=1, color='blue')
plt.plot(x_3, y_3, label='Locos por ofertas', linewidth=1, color='green')
plt.plot(x_4, y_4, label='Hombres con salas', linewidth=1, color='orange')

plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Volumen de Ingresos')
plt.grid(linewidth=0.5, linestyle='--', color='gray')
plt.ylim((0,120))

plt.legend(loc='best')
plt.show()
```

```python
# Show a heat map to understand the relation between weapons and target
mark_seg = ['M', 'HQR', 'LPO', 'HCS']
our_weap = ['Garantía', 'Virtual Reality', 'Web Móvil', 'Web PC', 'Flash!', 'Variedad', 'Descuentos', 'Envíos Gratis'] 

percentage = np.array([[0.9, 0.2, 0.2, 0.1],
                       [0.7, 0.0, 0.2, 0.9],
                       [0.3, 1.0, 0.6, 0.7],
                       [0.2, 0.4, 0.4, 0.3],
                       [0.4, 0.0, 1.0, 0.3],
                       [1.0, 0.6, 0.3, 0.7],
                       [0.8, 0.6, 0.9, 0.6],
                       [0.5, 0.1, 0.7, 0.3]])

# Heat map configuration
fig, ax = plt.subplots(figsize=(15,9))
im = ax.imshow(percentage, cmap="Reds")

ax.set_xticks(np.arange(len(mark_seg)))
ax.set_xticklabels(mark_seg)
ax.set_yticks(np.arange(len(our_weap)))
ax.set_yticklabels(our_weap)

# Loop over data dimensions and create text annotations.
for i in range(len(our_weap)):
    for j in range(len(mark_seg)):
        text = ax.text(j, i, round(percentage[i, j], 1),
                       ha="center", va="center", color="k", size=15)
        
ax.set_title('Relación Armas-Segmentos')
plt.show()
```
